We will drop those columns which have greater than 50% of the data missing

```python
# Assuming 'data' is your DataFrame with 1640 observations

# Calculate the percentage of missing values for each column
missing_percentage = (data.isnull().sum() / len(data)) * 100

# Filter out columns with more than 50% missing data
columns_to_drop = missing_percentage[missing_percentage > 50].index

# Drop the columns
data = data.drop(columns=columns_to_drop)
```

Now data contains only the columns with less than or equal to 50% missing data

```python
#data.isna().sum()
```

```python
data.shape
```

Filling missing values using measures of central tendency according to the datatype

```python
for x in data.columns:
    if data[x].dtype=='object' or data[x].dtype=='bool':
        data[x].fillna(data[x].mode()[0],inplace=True)
    elif data[x].dtype=='int64' or data[x].dtype=='float64':
        data[x].fillna(round(data[x].mean()),inplace=True)
```

Creating X and Y

```python
X=data.values[:,0:-1] 
Y=data.values[:,-1]
```

```python
X.shape
```

#### Converting Categorical Variables into numerical variables using Label encoder

```python
colname=[]
for x in data.columns:
    if data[x].dtype=='object':
        colname.append(x)
print(colname)
```

```python

from sklearn.preprocessing import LabelEncoder
 
le=LabelEncoder()
 
for x in colname:
    data[x]=le.fit_transform(data[x])
```

```python
# Now, you can proceed with scaling the data
scaler = StandardScaler()
scaler.fit(data_encoded)
X = scaler.transform(data_encoded)
```

```python
from sklearn.model_selection import train_test_split
 
#Split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, 
                                                    random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### Applying PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components= 0.95)   
X_train_new=pca.fit_transform(X_train)
X_test_new=pca.transform(X_test)
explained_variance=pca.explained_variance_ratio_
print(explained_variance)

```

```python
pca.n_components_
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
np.set_printoptions(suppress=True)

# create model object 
lm=LinearRegression()

# train model object
lm.fit(X_train_new,Y_train)

# print intercept and coefficients
print(lm.intercept_)
print(lm.coef_)


```

```python
#predict using model

Y_pred=lm.predict(X_test_new)
print(Y_pred)
```

```python
lm.score(X_train_new,Y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
 # ytest is used to cross check
r2=r2_score(Y_test,Y_pred) # unseen data
print("R-squared:",r2)
 
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))  # manually applied sqrt
print("RMSE:",rmse)
 
adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)  # formula of adj R2 shape(no of ind vars)
print("Adj R-square:",adjusted_r_squared)
```

values of R squared and adj R2 are close to each other and close to 1 hence it is ideal/good model

### Stochastic Gradient Descent

```python
from sklearn.linear_model import SGDRegressor

sgd_regressor = SGDRegressor(max_iter=1000, tol=1e-3)


sgd_regressor.fit(X_train_new, Y_train)

Y_pred = sgd_regressor.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
 
r2=r2_score(Y_test,Y_pred)
print("R-squared:",r2)
 
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred)) 
print("RMSE:",rmse)
 
adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)  # formula of adj R2 shape(shape-no of ind vars)
print("Adj R-square:",adjusted_r_squared)
```

```python
from sklearn.linear_model import SGDRegressor

# Define SGDRegressor with additional parameters
sgd_regressor = SGDRegressor(
    max_iter=1000,  # Maximum number of iterations
    tol=1e-3,       # Tolerance for stopping criteria
    penalty='l2',   # Regularization term (l2 penalty)
    alpha=0.0001,   # Regularization strength
    learning_rate='invscaling',  # Learning rate schedule
    eta0=0.01,      # Initial learning rate
    verbose=1       # Verbosity level (1 for default verbosity)
)

# Fit the model
sgd_regressor.fit(X_train_new, Y_train)

# Make predictions
Y_pred = sgd_regressor.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
 
r2=r2_score(Y_test,Y_pred)
print("R-squared:",r2)
 
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred)) 
print("RMSE:",rmse)
 
adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)  # formula of adj R2 shape(shape-no of ind vars)
print("Adj R-square:",adjusted_r_squared)
```
